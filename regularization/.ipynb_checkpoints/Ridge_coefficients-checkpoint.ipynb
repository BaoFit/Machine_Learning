# 绘制Ridge回归惩罚项参数对学习器的影响

<p> item:每种颜色代表系数矢量的不同特征，并且根据正则化参数进行显示。
<p> conclusion：</p>
<p> 当alpha很大时，正则化效应将主导平方损失函数，并且系数趋于零。在路径的末尾，随着alpha趋于零且解趋于普通的最小二乘，系数表现出较大的振荡
<p> 建模过程中，有必要以使两者之间保持平衡的方式调整alpha。

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# X 时 10x10 的Hilbert matrix
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis])
y = np.ones(10)

n_alphas = 200
#构造200个从10的-10次方到10的-2次方的等比数列
alphas = np.logspace(-10, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)

#获得当前的子图
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

```python
# X 时 10x10 的Hilbert matrix
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis])
y = np.ones(10)+1

n_alphas = 200
#构造200个从10的-10次方到10的-2次方的等比数列
alphas = np.logspace(-10, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)

#获得当前的子图
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

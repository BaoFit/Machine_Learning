# L1-normal/L2-normal
为了更加直观的理解L1正则化和L2正则化，简单的画个图感受一下
<p>item：正则化参数的选择

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

如下图所示,J0=(x-1)^2 表示损失函数，L1正则化惩罚项|x|,正则化系数越大，最小值的解越接近0

```python
x = np.linspace(-1,2,50)
y1 = x*x - 2*x + 1
y2 = abs(x)
plt.plot(x,y1+ 0.5*y2,label=r'$(x-1)^2+0.5|x|$')
plt.plot(x,y1+ y2,label=r'$(x-1)^2+|x|$')
plt.plot(x,y1+ 2*y2,label=r'$(x-1)^2+2|x|$')
plt.annotate('min-num',(0,1),xytext=(0.45, 0.35), textcoords='axes fraction',
            arrowprops=dict(facecolor='green', shrink=0.05, headwidth=4,width=0.1),
            fontsize=10,horizontalalignment='center', verticalalignment='top')

plt.annotate('min-num',(0.8,0.4),xytext=(0.75, 0.65), textcoords='axes fraction',
            arrowprops=dict(facecolor='blue', shrink=0.05, headwidth=4,width=0.1),
            fontsize=10,horizontalalignment='center', verticalalignment='top')

plt.annotate('min-num',(0.6,0.76),xytext=(0.55, 0.55), textcoords='axes fraction',
            arrowprops=dict(facecolor='yellow', shrink=0.05, headwidth=4,width=0.1),
            fontsize=10,horizontalalignment='center', verticalalignment='top')

plt.legend(loc='best')
plt.xlabel(r'x')
plt.ylabel(r'$(x-1)^2+\lambda|x|$')
plt.title(r'$L_1$-normal with different $\lambda$')
plt.show()
```
